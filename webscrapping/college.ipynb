```python
def fetch_page_data(url):
    response=requests.get(url)
    soup=BeautifulSoup(response.content,'html.parser')
    return soup
```

```python
def extract_college_names(soup):
     college_names = []
     links = soup.find_all('div', attrs={'class': 'titleSection'})
     for link in links:
          title_text = link.find('h3').get_text(strip=True)
          college_names.append(title_text)
     return college_names
```

```python
all_college_names = []
base_url = "https://www.collegedekho.com/btech-colleges-in-india/?page="
num_pages = 6 

for page in range(2, num_pages + 1):
    url = f"{base_url}{page}"
    print(f"Fetching data from: {url}")
    
    try:
        soup = fetch_page_data(url)
        college_names = extract_college_names(soup)
        all_college_names.extend(college_names)
       
    except requests.exceptions.RequestException as e:
        print(f"Error fetching page {page}: {e}")
        continue
```

```python
for name in all_college_names:
    print(name)
```
