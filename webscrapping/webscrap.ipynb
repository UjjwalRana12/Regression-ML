```python
from bs4 import BeautifulSoup
```

```python
import requests
import pandas as pd
```

```python
URL =  'https://www.flipkart.com/clothing-and-accessories/pr?sid=clo&otracker=categorytree&p%5B%5D=facets.ideal_for%255B%255D%3DMen&otracker=nmenu_sub_Men_0_Clothing'
```

```python
# Headers for request
HEADERS = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36 Edg/126.0.0.0',
    'Accept-Language': 'en-US, en;q=0.5'
}
```

```python
webpage=requests.get(URL,headers=HEADERS)
```

```python
webpage
```

```python
soup=BeautifulSoup(webpage.content,"html.parser")
```

```python
soup
```

```python
links= soup.find_all("a",attrs={'class':'rPDeLR'})
```

```python
links
```

```python
links[0]
```

```python
link=links[0].get('href')
```

```python
link
```

```python
product_list="https://flipkart.com" + link
```

```python
product_list
```

```python
new_webpage=requests.get(product_list,headers=HEADERS)
```

```python
new_webpage
```
