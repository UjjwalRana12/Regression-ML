```python
from sklearn.preprocessing import StandardScaler
sc_x =StandardScaler()
sc_y =StandardScaler()
x=sc_x.fit_transform(x)
y=sc_y.fit_transform(y)
```

Creating two separate StandardScaler objects for x and y is necessary when you need to scale both the input features (x) and the target variable (y). This is especially important when the ranges of x and y are different and need to be scaled independently.

Why Two StandardScaler Objects?
Independent Scaling: x and y are generally on different scales, and they need to be scaled independently. Each StandardScaler object maintains its own mean and standard deviation, which are used to scale the data

When the ranges of x and y are different, it means that the numerical values of the input features (x) and the target variable (y) vary on different scales. For example, the features might range from 1 to 10, while the target variable might range from 1.5 to 91.5, as shown in the previous example. This difference in scale can affect the performance of machine learning models, particularly those that rely on distance metrics or optimization algorithms that are sensitive to the scale of the data.
