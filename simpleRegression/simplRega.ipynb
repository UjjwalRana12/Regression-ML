```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Salary_Data.csv')
x= dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

now i have to enter what i want to predict therefore i take x_test and x test is the testing salary and y test is the real salary which we check after predicting therefore i save it in y_predict so that i compare it to y_test 

```python
regressor.predict(x_test)
y_pred=regressor.predict(x_test)
```

now we would show this in prpoer ways lets visualize our data for this we use matplotlib scatter in which i put params x train and ytrain beacuse i have to give years of experience and earnings and graph of them which will depicts 
OR WE CAN SAY THAT WE ARE VISUALIZING TRAINING SET
THEN WE HAVE TO SHOWS REGRESSION LINE SO WE CAN SHOW THAT HOW MUCH OUR POINTS THAT REPRESENT DATA LIES ON A LINE

```python
plt.scatter(x_train,y_train , color='red')
plt.plot(x_train,regressor.predict(x_train),color='blue')
plt.title('salary vs experience(TRAINING SET)')
plt.xlabel('exp')
plt.ylabel('salary')
plt.show()
```

```python
plt.scatter(x_test,y_test , color='red')
plt.plot(x_train,regressor.predict(x_train),color='blue')
plt.title('salary vs experience(TEST SET)')
plt.xlabel('exp')
plt.ylabel('salary')
plt.show()
```
