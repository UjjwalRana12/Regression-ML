```python
from sklearn.impute import SimpleImputer
```

```python
imp_mode=SimpleImputer(missing_values=np.nan,strategy="most_frequent")
```

```python
df_imputed=pd.DataFrame(imp_mode.fit_transform(data))
```

```python
sns.heatmap(df_imputed.isnull())
```

```python
df_imputed.columns=data.columns
```

```python
df_imputed=df_imputed.drop('id',axis=1)
```

```python

df_imputed['age'] = df_imputed['age'].astype(str).replace('51.48337595907928', '55.0').astype(float)
df_imputed['bp'] = df_imputed['bp'].astype(str).replace('76.46907216494846', '76.0').astype(float)
df_imputed['sg'] = df_imputed['sg'].astype(str).replace('1.0174079320113314', '1.017').astype(float)
```

```python
df_imputed['age']=df_imputed['age'].astype(int)
df_imputed['bp']=df_imputed['bp'].astype(int)
df_imputed['sg']=df_imputed['sg'].astype(float)
df_imputed['al']=df_imputed['al'].astype(int)
df_imputed['su']=df_imputed['su'].astype(int)

```

```python
df_imputed.info()
```

```python
df_imputed['rbc'].unique()
```

```python
df_imputed['rbc'] = df_imputed['rbc'].map({'normal': 1, 'abnormal': 0})
df_imputed['pc'] = df_imputed['pc'].map({'normal': 1, 'abnormal': 0})
df_imputed['pcc'] = df_imputed['pcc'].map({'present': 1, 'notpresent': 0})
df_imputed['ba'] = df_imputed['ba'].map({'present': 1, 'notpresent': 0})
```

```python
df_imputed['rbc']=df_imputed['rbc'].astype(int)
df_imputed['pc']=df_imputed['pc'].astype(int)
df_imputed['pcc']=df_imputed['pcc'].astype(int)
df_imputed['ba']=df_imputed['ba'].astype(int)
df_imputed['bgr']=df_imputed['bgr'].astype(float)
df_imputed['bu']=df_imputed['bu'].astype(float)
```
